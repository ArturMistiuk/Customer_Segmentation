# Changing Columns

```python
pd.to_datetime(df["Dt_Customer"], format="%d-%m-%Y").max()
```

```python
df["Children"] = df["Teenhome"] + df["Kidhome"]
```

```python
df["Age"] = 2014 - df["Year_Birth"]
```

```python
df["Membership"] = pd.to_datetime("2014-06-30") - pd.to_datetime(df["Dt_Customer"], format="%d-%m-%Y")
df["Membership"] = df["Membership"].astype("str")
df["Membership"] = df["Membership"].str.split()
df["Membership"] = df["Membership"].apply(lambda x: x[0]).astype("int")
```

```python
df["Partner_Status"] = df["Marital_Status"].replace({"Single": 0, "Divorced": 0, "Widow": 0, "Alone": 0, "Absurd": 0, "YOLO": 0,
                                                   "Together": 1, "Married": 1}).astype("int")
```

```python
df = df.rename(columns={
    "MntFruits": "Fruits", "MntMeatProducts": "Meat", "MntFishProducts": "Fish",
    "MntSweetProducts": "Sweets", "MntWines": "Wines", "MntGoldProds": "Gold",
    "NumStorePurchases": "Store_Purchases", "NumWebPurchases": "Web_Purchases"
})
```

```python
df["Is_Parent"] = np.where(df["Children"] > 0, 1, 0)
```

```python
df["Graduated"] = df["Education"].replace({"Basic": 0, "2n Cycle": 0, "Master": 1, "PhD": 1, "Graduation": 1}).astype("int")
```

```python
df["Spending"] = df["Wines"] + df["Fruits"] + df["Meat"] + df["Fish"] + df["Sweets"] + df["Gold"]
```

# KMeans Clustering 

```python
import os
os.environ['OMP_NUM_THREADS'] = '9'
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
```

```python
X = df.copy()
```

```python
scaler = StandardScaler()
cluster = KElbowVisualizer(KMeans(n_init=10))
pipe = make_pipeline(scaler, cluster)

pipe.fit(X)
```

```python
scaler = StandardScaler()
cluster = KMeans(n_init=10, n_clusters=5)
pipe = make_pipeline(scaler, cluster)

pipe.fit(X)
```

```python
cluster.labels_
```

```python
df["Cluster"] = cluster.labels_
```

```python
other["Cluster"] = cluster.labels_
```
